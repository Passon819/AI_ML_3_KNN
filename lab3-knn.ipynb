```python
from sklearn.neighbors import KNeighborsClassifier

#split data into train(1000 ภาพแรก)/test(797 ภาพหลัง)
X_train = digits.data[:1000] # 1000 ภาพแรก (0-999)
Y_train = digits.target[:1000] # 1000 label แรก
X_test = digits.data[1000:] # 797 ภาพหลัง (1000-1797)
Y_test = digits.target[1000:] 

clf = KNeighborsClassifier(n_neighbors = 1) # n_neighbors = 1 (1NN)
clf.fit(X_train, Y_train) # ฝึก

pred = clf.predict(X_test) # ทำนาย
n_correct = sum(pred == Y_test) #นับ true นับจำนวนภาพที่คอมทำนายถูก
acc = n_correct/len(Y_test) # accuracy (ความแม่นยำ) = #ตอบถูก/#ทั้งหมด
acc
test_error = 1 - acc
test_error #สำคัญกว่า train error
```
